```python
def ff_whiff_percent(playerid):
    playerdata = statcast_pitcher("2020-01-01","2020-09-28",playerid)
    playerdata["swing"] = np.where((playerdata["description"].isin(["foul","foul_tip","hit_into_play","hit_into_play_no_out","hit_into_play_score","swinging_strike","swinging_strike_blocked"])),1,0)
    playerdata["swstr"] = np.where((playerdata["description"].isin(["swinging_strike","swinging_strike_blocked","foul_tip"])),1,0)
    player_fastballs = playerdata[playerdata["pitch_type"]=="FF"]
    whiff_percent = player_fastballs["swstr"].sum()/player_fastballs["swing"].sum()
    return(whiff_percent)
```

```python
def whiff_percent(playerid):
    playerdata = statcast_pitcher("2020-01-01","2020-09-28",playerid)
    playerdata["swing"] = np.where((playerdata["description"].isin(["foul","foul_tip","hit_into_play","hit_into_play_no_out","hit_into_play_score","swinging_strike","swinging_strike_blocked"])),1,0)
    playerdata["swstr"] = np.where((playerdata["description"].isin(["swinging_strike","swinging_strike_blocked","foul_tip"])),1,0)
    whiff_percent = playerdata["swstr"].sum()/playerdata["swing"].sum()
    return(whiff_percent)
```

```python
spin_data = pd.read_csv("spin-direction-pitches.csv")
```

```python
spin_data.head(5)
```

```python
playerid_list = spin_data["player_id"].values.tolist()
```

```python
fastball_whiffs = []
for ids in playerid_list:
    ff_whiff = ff_whiff_percent(ids)
    fastball_whiffs.append(ff_whiff)
```

```python
spin_data["FF_Whiff%"] = fastball_whiffs
```

```python
spin_data["Spin_Discounted"] = spin_data["spin_rate"]*spin_data["active_spin"]
```

```python
print(spin_data["FF_Whiff%"].corr(spin_data["spin_rate"]))
```

```python
spin_data["Velo_Percentile"] = spin_data.release_speed.rank(pct=True)
```

```python
spin_data["Spin_Percentile"] = spin_data.spin_rate.rank(pct=True)
```

```python
spin_data["Velo*Spin"] = spin_data["Velo_Percentile"]*spin_data["Spin_Percentile"]
```

```python
print(spin_data["FF_Whiff%"].corr(spin_data["Velo*Spin"]))
```

```python
spin_data["Active_Percentile"] = spin_data.active_spin.rank(pct=True)
```

```python
spin_data["Velo*Spin*Active"] = spin_data["Velo_Percentile"]*spin_data["Spin_Percentile"]*spin_data["Active_Percentile"]
```

```python
print(spin_data["FF_Whiff%"].corr(spin_data["Velo*Spin*Active"]))
```

```python
spin_data["Velo*Active"] = spin_data["Velo_Percentile"]*spin_data["Active_Percentile"]
```

```python
print(spin_data["FF_Whiff%"].corr(spin_data["Velo*Active"]))
```
